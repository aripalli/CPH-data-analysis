##### Write about your choices in data cleaning and preprocessing

#### Nationality data set

The nationality data is pretty clean, however, we want to get remove every useless data in the .csv files for the website. 

For the nationality dataset, we wanted to find the ratio between a specific nationality and all foreign nationality both in every district and for all Copenhagen. We feel like inspecting the ratio will also give a good sense of how distinct nations are changing through the years for both the unique district and all of Copenhagen. 

Example: (Not true values!)

*Number of Icelandic people living in Østerbro the year 2015 was 100, however, the total number of foreigners living in Østerbro was 1000. So, Icelandic people counted as 100/1000 = 10% of the foreigners living in Østerbro.*

*Number of foreigners living in Copenhagen the year 2015 was 100 however the total number of Danes living in Copenhagen was 1000. So, foreigners in Copenhagen counted as 100/1000 = 10% of the people living in Copenhagen.*

We wanted also to change the numerical value of districts and nations to their names to give a better understanding of the visualization. Unfortunately, the decoder for the nation number to nation name had only the Danish name of the country. However, it is always fun to learn a new danish word :). Also, we wanted, to sum up, all of the districts each year for every nation to be able to see the number of a specific nation living in Copenhagen, not just in a district in Copenhagen.

#### Socio economic data set
The socio data is not as clean as the nationality data. We had to do different kinds of preprocessing for different tasks. First of all a lot of values were missing. These values were set to the mean of the given feature for the given year. Some features were completly missing for some years. These were just removed. When doing elastic net the data was furthermore standardized. This was to ensure that when we tried to interpret the coefficients, their size made sense. A lot of the features were both found in a percent version but also a number version. We decided to remove all of the ones that was just a number, since the size of the rode would influence this.
We noticed that for the year 2014 too many important values were missing. Therefore we decided to not work with this year.

#### Nationality data set
Below is a brief overview of the data. It is loaded into a pandas data frame and the header is printed after the county names have been added. There is also a plot showing the total amount of foreginers for 2015. Denmark has been removed from the data.

```python
data_nat = pd.read_csv('../data/befkbhalderstatkode.csv')
decoder=pd.read_csv('../data/Statcode_decoder.csv')
data_nat['Country']=0
nationality_code=list(set(data_nat.STATKODE))
Years_nat=list(set(data_nat.AAR))

for code in nationality_code:
    country=decoder.query('STATKODE==@code')['COUNTRY'].tolist()[0]
    data_nat.loc[data_nat.STATKODE == code, ['Country']] = country
    
people=[]
subdata=data_nat.query('AAR==2015')
for code in nationality_code:
    #Ignore Denmark
    if code==5100:
        continue
    people.append(subdata.query('STATKODE==@code').PERSONER.sum())

data_nat.head()
```

```python
import matplotlib.pyplot as plt
from operator import itemgetter
import numpy as np
%matplotlib inline
#Look at 2015 as an example
subdata=data_nat.query('AAR==2015')

#Count people
people=[]
for code in nationality_code:
    people.append(subdata.query('STATKODE==@code').PERSONER.sum())
#ADd countries in the right order.
countries=[]
for stat in nationality_code:
    countries.append(decoder.query('STATKODE==@stat').COUNTRY.tolist()[0].decode('utf-8'))

#Counter of how many people live in the countries
count_people=zip(countries,people)
count_people_sorted=sorted(count_people,key=itemgetter(1),reverse=True)
countries=zip(*count_people_sorted)[0]
people=zip(*count_people_sorted)[1]
#Plot top 100
top_num=100
y_pos = np.arange(len(countries))
#Plot it
plt.bar(y_pos[1:top_num], people[1:top_num], align='center', alpha=0.5)
plt.xticks(y_pos[1:top_num], countries[1:top_num],rotation='vertical')
plt.ylabel('Number of people')
plt.xlabel('Countries')
plt.title('Distribution of foreigners in 2015. Top %d'%top_num)
axes = plt.gca()
axes.set_xlim([0,top_num])
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.show()
```

#### Describe which machine learning tools you use and why the tools you've chosen are right for the problem you're solving.

##### Regression
We used 2 regression methods. We used elastic net and random forest. Random forest is generally a really good method. It furthermore provides a measure of how important features are, which is ideal for what we want to do!
However when there are a lot of features compared to observations. Which is the case when we only look at a given year. Then elastic net shines. Since it has a nice way of doing dimensionality reductions. Elastic net is thus used as comparison. As we standardize the data, we can also look at the sizes of the coefficients, to also here get an estimate of the important features.

##### Clustering
We used K-means to cluster the districts together. K-means is a simple and well performing method which was why we choose it. We also tried DBSCAN but it performed wores, so we stuck to K-means

#### Talk about your model selection. How did you split the data in to test/training. Did you use cross validation?

To get a true measure of our error we did 10 fold cross validation. For each fold we build the model and tested on on the test set. We then stored the error. We then determined the true performance as the average of all the 10 errors. We split the data 10 ways using the sklearn Kfold function. The variables in the models was determined based on what we thought made sense. Initially we set up another 10 fold outer cross validation loop such that we also could determine the optimal parameters. However because on the size of the data that took way too long! And we quickly saw that the choice of parameters did not affect the errors a lot.

For clustering we made an error plot to determine the knee point to get an optimal number of clusters. We decided to go with 4. As this is an unsupervised method cross validation can not be done.

#### Explain the model performance. How did you measure it? Are your results what you expected?

As mentioned above we determined the model performance as the average error after a 10 fold cross validation. This was to ensure that we did not test and train the model on the same data set.
