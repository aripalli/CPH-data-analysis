## Motivation

#### What is your dataset?

We have two datasets, both from the [Open Copenhagen data](http://data.kk.dk/) website. The website contains all open data from the Copenhagen kummune (e. Municipality). We downloaded the files directly from the website.

The first dataset is the population by year, district, age, and citizenship. You can find the dataset **[here](http://data.kk.dk/dataset/befolkningen-efter-ar-bydel-alder-og-statsborgerskab)**. We will call this dataset nationality through the text.

The second dataset is the socioeconomic dataset. This is a very large dataset with a lot of information. You can find the dataset **[here](http://data.kk.dk/dataset/samlede-socio-data-kbh)**. We will call this dataset socioeconomic through the text.

#### Why did you choose this/these particular dataset(s)?

We wanted to know whether it is possible to predict where and how international and Danes live in Copenhagen. Furthermore, see if some areas show certain trends and try to explain why there is a trend.

The first dataset is ideal to make the visualizing of foreigners in Copenhagen. It contains information on the population of every nation from 1992 to 2015 and also splits Copenhagen ten districts (and one unknown).

The second dataset is ideal to use for the prediction has it contains a lot of features which could be useful and interesting to use to predict where people live and if there is any trend (e.g. If through the years there is more trend of Danes living in a specific district or more of international people).
